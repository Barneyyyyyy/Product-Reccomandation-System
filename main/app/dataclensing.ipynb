```python
import pandas as pd
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler
#garbage collector
import gc
gc.enable() #enabling collection
```

```python
order_products_train_df = pd.read_csv('../../CSVs/order_products__train.csv')
orders_df = pd.read_csv('../../CSVs/orders.csv')
products_df = pd.read_csv('../../CSVs/products.csv')
```

```python
# Define a function for common data cleaning steps
def clean_data(df, apply_zscore=False, apply_normalization=False):
    # Inspect the Data
    print(df.head())
    print(df.info())
    print(df.describe())

    # Handle Missing Values
    print(df.isnull().sum())
    # Decide on a strategy and apply it
    # For example, fill missing values with a placeholder or handle them as appropriate for text data
    df.fillna('unknown', inplace=True)

    # Data Type Conversion
    # (No specific example provided, as it depends on the columns in each dataframe)

    # Remove Duplicates
    df.drop_duplicates(inplace=True)

    # Optional: Outlier Detection and Handling
    if apply_zscore:
        z_scores = zscore(df.select_dtypes(include='number'))
        df = df[(z_scores < 3).all(axis=1)]

    # Optional: Normalization/Standardization
    if apply_normalization:
        scaler = MinMaxScaler()
        df = pd.DataFrame(scaler.fit_transform(df.select_dtypes(include='number')), columns=df.columns)

    return df
```

```python
# Apply cleaning to each dataset
order_products_train = clean_data(order_products_train_df,apply_zscore=True,apply_normalization=True)
orders = clean_data(orders_df)
products = clean_data(products_df)
```
